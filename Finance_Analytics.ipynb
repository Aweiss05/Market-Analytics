```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# download ticker for one day
ticker = yf.download(tickers = "SPY", period = "1d")

# get the last close of the stock price
latest_close = ticker["Close"].iloc[-1]

t0 = latest_close # starting stock price of SPY
ear = 0.10 # expected annual return of SPY
sigma = 0.1123 # average annual volitility
t = 1 # years
n = 252 # trading days in a year, steps
dt = t / n # step size of the time

z = np.random.normal(0, 1, n)

S = np.zeros(n)
S[0] = t0
for t in range(1, n):
  S[t] = S[t-1] * np.exp((ear - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * z[t])

# above represents the formula for Geometric Brownian Motion (GBM for short)

plt.plot(S)
plt.title("SPY Simulated Stock Price using GBM")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
```
