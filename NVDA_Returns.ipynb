```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download NVDA historical data
nvda = yf.download('NVDA', '2024-01-01')

# Calculate 50-day moving average and standard deviation
nvda['ma_50'] = nvda['Close'].rolling(50).mean()
nvda['sd_50'] = nvda['Close'].rolling(50).std()

# Calculate upper and lower Bollinger Bands
nvda['upper_band'] = nvda['ma_50'] + 2 * nvda['sd_50']
nvda['lower_band'] = nvda['ma_50'] - 2 * nvda['sd_50']

# Plot only the Bollinger Bands
plt.figure(figsize=(12, 6))
plt.plot(nvda.index, nvda['ma_50'], label='50-Day MA', color='blue')
plt.plot(nvda.index, nvda['upper_band'], label='Upper Band', color='green', linestyle='--')
plt.plot(nvda.index, nvda['lower_band'], label='Lower Band', color='red', linestyle='--')
plt.fill_between(nvda.index, nvda['lower_band'], nvda['upper_band'], color='gray', alpha=0.2)

plt.title('NVDA 50-Day Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()


```
